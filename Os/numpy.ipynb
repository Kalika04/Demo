Array Attribute

```python
arr=np.arange(60).reshape(12,5)   ## 12 rows and 5 columns
arr
```

```python
print("array shape:-",arr.shape )
print("array size:-",arr.size)
print("array datatype:- ",arr.dtype)
print("array ndim:- ",arr.ndim)  ## dimension of an array
```

```python
# 2D array into 1D array

arr = arr.flatten()
print("array shape:-",arr.shape )
print("array size:-",arr.size)
print("array datatype:- ",arr.dtype)
print("array ndim:- ",arr.ndim)  
```
