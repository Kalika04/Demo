```python
# pie chart

x=np.array([1,2,3,4])
labels = ['a','b','c','d']
plt.pie(x, labels = labels)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels = ['a','b','c','d']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = labels,explode = myexplode)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels = ['a','b','c','d']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = labels,explode = myexplode, shadow = True )
plt.show()
```

```python
colour = ['red','green','blue','yellow']
x=np.array([1,2,3,4])
labels = ['a','b','c','d']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = labels,explode = myexplode, shadow = True, colors = colour)
plt.show()
```

```python
colour = ['red','green','blue','yellow']
x=np.array([1,2,3,4])
labels = ['a','b','c','d']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = labels,explode = myexplode, shadow = True, colors = colour, autopct = '%1.1f%%')
plt.show()
```

```python
colour = ['red','green','blue','yellow']
x=np.array([1,2,3,4])
labels = ['a','b','c','d']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = labels,explode = myexplode, shadow = True, colors = colour, autopct = '%1.1f%%')
plt.legend('my pie chart')
plt.show()
```
